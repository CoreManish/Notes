```python
import seaborn as sns
```

# histogram : used to show continuous data and their frequency
* it automatically make interval on x axis called class interval

```python
# marks of students
x=[95,98.5,80,34,55.5,61,66,5,45]

sns.histplot(x=x, kde=True)
```

```python
# Number of bus in different cities
# it is dicrete value, so histogram will not give any proper idea
x = [90, 85, 95,90,54,85,60,85]

# Create a bar plot
sns.histplot(x=x)
```

```python
x = ['Male', 'Male', 'Other', 'Male', 'Female','Female']
# Create a bar plot
sns.histplot(x=x)
```

# barplot : used to show categorical data and their frequency

```python
x = ['Male', 'Female', 'Other']
y = [2, 4, 6]
cols = ['red' if c>3 else 'green' for c in y]
# Create a bar plot
sns.barplot(x=x, y=y, palette=cols)
```

```python
#marks
x = [90, 85, 95]
y = [2, 4, 6]

# Create a bar plot
sns.barplot(x=x, y=y)
```
