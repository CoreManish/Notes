```python
import numpy as np

def is_symmetric(X):
  if X.shape[0]==X.shape[-1]:
    ans=1
    X_t=np.transpose(X)
    for i in range(X.shape[0]):
      for j in range(X.shape[-1]):
        if X[i][j] != X_t[i][j]:
          ans=0
  else:
    ans=0
  return ans
```

```python
import numpy as np

def update_weight(X,y,n,alpha):
  w=np.zeros(X.shape[-1])
  for k in range(n):
    for index in range(X.shape[0]):
      if ( y[index] * (np.transpose(w)@ X[index]) ) <=0:
        w= w + alpha * y[index] * X[index]
  return 2*w
```
