# Normal distribution

```python
from scipy.stats import norm
```

```python
norm.pdf(5,3,1)
```

```python
norm.cdf(5,3,1)
```

```python
norm.cdf(5,3,1)
```

### Generate random sample

```python
X = norm.rvs(0,1,100)
X
```

```python
sns.histplot(X)
```

```python
Y = norm.pdf(X,0,1)
Y
```

```python
import seaborn as sns
sns.scatterplot(x=X,y=Y)
```

# POISSON DISTRIBUTION

```python
from scipy.stats import poisson
```

### Find PMF

```python
# Find probability that 7 times event will happen
poisson.pmf(7,mu=5)
```

```python
# Find probability for multiple values
poisson.pmf([0,3,5,23,7],mu=5)
```

### Find CDF

```python
poisson.cdf(7,mu=5)
```

```python
poisson.cdf([0,3,5,23,7], mu=5)
```

### Generate samples from poisson random variable

```python
# Generate sample from poisson random variable
X=poisson.rvs(mu=12,size = 1000)
```

```python
Y = poisson.pmf(X,mu=5)
```

```python
import seaborn as sns
sns.scatterplot(x=X,y=Y)
```
